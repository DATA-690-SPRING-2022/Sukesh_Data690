```python
import pandas as pd
import plotly
import plotly.express as px
```

```python
Data_url = 'https://raw.githubusercontent.com/Sukesh-bitragunta/Sukesh_Data690/main/world_development_explorer/GDP.csv'

df = pd.read_csv(Data_url)
```

All the developed contries suchs as USA, England, Australia have a good finacial sector and that helped them have a steady economy and growth, we witnessed this in the pandemic itself.
```python
df_2000_GDP = df.query("Year == 2000").query("indicator == 'BM.KLT.DINV.WD.GD.ZS'")
df_2000_GDP
```

The below pie charts indicates the direct investement of foriegn countries to the respective countries and what percent of their GPD consits of this foreign direct investments.

In 2000
Malaysia had more than 36 pecent of foriegn direct investements that directly contributes their GDP in year 2000

Russian Federation had around 20.8 percent and the difference between Australia and Japan is ver minimal that they are almost equal. China, Georgia, Thailand, India, Sri Lanka had less than 10 percent in 2000.

```python
fig = px.pie(df_2000_GDP, values='value', names='Country Name', title='2000 total GDP per capita Comparision between given nations.',template=list(plotly.io.templates.keys())[5])
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(showlegend=True)

fig.show()
```

```python
df_2018_GDP = df.query("Year == 2018").query("indicator == 'BM.KLT.DINV.WD.GD.ZS'")
df_2018_GDP
```

```python
fig = px.pie(df_2018_GDP, values='value', names='Country Name', title='2018 total GDP per capita Comparision between given nations.',template=list(plotly.io.templates.keys())[5])
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(showlegend=True)

fig.show()
```
